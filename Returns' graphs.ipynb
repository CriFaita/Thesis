```python
import yfinance as yf
import pandas as pd
import sklearn
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.metrics import mean_squared_error
from math import sqrt

plt.rcParams["figure.figsize"] = [20, 25]
```

```python
# download all stocks' values and compute returns

apl = yf.Ticker("AAPL")
apple = apl.history(start="2015-01-05", end="2021-09-01", interval="1d")
returns_Apple = apple['Close'] + apple['Dividends']

ggl = yf.Ticker("GOOG")
google = ggl.history(start="2015-01-05", end="2021-09-01", interval="1d")
returns_Alphabet = google['Close'] + google['Dividends']

amc = yf.Ticker("AMC")
Amc = amc.history(start="2015-01-05", end="2021-09-01", interval="1d")
returns_AMC = Amc['Close'] + Amc['Dividends']

btc = yf.Ticker("BTC-USD")
Btc = btc.history(start="2015-01-05", end="2021-09-01", interval="1d")
returns_BTC = Btc['Close'] + Btc['Dividends']

fb = yf.Ticker("FB")
Fb = fb.history(start="2015-01-05", end="2021-09-01", interval="1d")
returns_FB = Fb['Close'] + Fb['Dividends']

gme = yf.Ticker("GME")
Gme = gme.history(start="2015-01-05", end="2021-09-01", interval="1d")
returns_GME = Gme['Close'] + Gme['Dividends']

qqq = yf.Ticker("QQQ")
Qqq = qqq.history(start="2015-01-05", end="2021-09-01", interval="1d")
returns_QQQ = Qqq['Close'] + Qqq['Dividends']

jnj = yf.Ticker("JNJ")
Jnj = jnj.history(start="2015-01-05", end="2021-09-01", interval="1d")
returns_JNJ = Jnj['Close'] + Jnj['Dividends']

msft = yf.Ticker("MSFT")
Msft= msft.history(start="2015-01-05", end="2021-09-01", interval="1d")
returns_MSFT = Msft['Close'] + Msft['Dividends']

nflx = yf.Ticker("NFLX")
Nflx = nflx.history(start="2015-01-05", end="2021-09-01", interval="1d")
returns_NFLX = Nflx['Close'] + Nflx['Dividends']

pfe = yf.Ticker("PFE")
Pfe = pfe.history(start="2015-01-05", end="2021-09-01", interval="1d")
returns_PFE = Pfe['Close'] + Pfe['Dividends']

wen = yf.Ticker("WEN")
Wen = wen.history(start="2015-01-05", end="2021-09-01", interval="1d")
returns_WEN = Wen['Close'] + Wen['Dividends']
```

```python
# create graphs

fig, axs = plt.subplots(12)
fig.suptitle('Fig.11: Returns', y=0.905, fontsize=20, fontstyle = 'italic')

#Alphabet
axs[0].plot(returns_Alphabet)
axs[0].set_title('Alphabet')
# axs[0].set_xlabel('Time')
# axs[0].set_ylabel('Returns')

#AMC
axs[1].plot(returns_AMC)
axs[1].set_title('AMC')

# Apple
axs[2].plot(returns_Apple)
axs[2].set_title('Apple')

# Bitcoin
axs[3].plot(returns_BTC)
axs[3].set_title('Bitcoin')

# FaceBook
axs[4].plot(returns_FB)
axs[4].set_title('FaceBook')

# GameStop
axs[5].plot(returns_GME)
axs[5].set_title('GameStop')

# QQQ
axs[6].plot(returns_QQQ)
axs[6].set_title('QQQ ETF')

# JNJ
axs[7].plot(returns_JNJ)
axs[7].set_title('Johnson & Johnson')

# MSFT
axs[8].plot(returns_MSFT)
axs[8].set_title('Microsoft')

# Netflix
axs[9].plot(returns_NFLX)
axs[9].set_title('Netflix')

# Pfizer
axs[10].plot(returns_PFE)
axs[10].set_title('Pfizer')

# Wendy's
axs[11].plot(returns_WEN)
axs[11].set_title("Wendy's")
```
